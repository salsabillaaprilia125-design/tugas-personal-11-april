Implementasi Hash Table Sederhana

```python
class HashTable:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size

    # Fungsi Hash
    def hash_function(self, key):
        return key % self.size

    # Insert dengan Linear Probing
    def insert(self, key):
        index = self.hash_function(key)

        while self.table[index] is not None:
            index = (index + 1) % self.size

        self.table[index] = key

    # Menampilkan isi Hash Table
    def display(self):
        print("\nIsi Hash Table")
        print("-" * 30)

        for i in range(self.size):
            if self.table[i] is None:
                print(f"Index {i} : Empty")
            else:
                print(f"Index {i} : {self.table[i]}")


# ===========================
# Program Utama
# ===========================

hash_table = HashTable(10)

data = [25, 36, 47, 58, 69, 80]

for item in data:
    hash_table.insert(item)

hash_table.display()
```

Mengurutkan Data Mahasiswa Menggunakan Quick Sort


```python
# ==================================
# Fungsi Quick Sort
# ==================================

def quick_sort(data):

    if len(data) <= 1:
        return data

    # memilih pivot (elemen tengah)
    pivot = data[len(data) // 2]

    left = []
    middle = []
    right = []

    for item in data:

        if item < pivot:
            left.append(item)

        elif item == pivot:
            middle.append(item)

        else:
            right.append(item)

    return quick_sort(left) + middle + quick_sort(right)


# ==================================
# Program Utama
# ==================================

nilai = [78, 65, 90, 88, 55, 72, 95, 60, 84, 70]

print("Data Awal")
print(nilai)

hasil = quick_sort(nilai)

print("\nData Setelah Quick Sort")
print(hasil)

jumlah_data = len(hasil)
nilai_min = min(hasil)
nilai_max = max(hasil)
rata_rata = sum(hasil) / jumlah_data

print("\nStatistik Data")
print("----------------------")
print("Jumlah Data :", jumlah_data)
print("Nilai Minimum :", nilai_min)
print("Nilai Maksimum :", nilai_max)
print("Nilai Rata-rata :", round(rata_rata, 2))
```

Perbandingan Bubble Sort dan Merge Sort

```python
import time

# ==========================================
# Bubble Sort
# ==========================================

def bubble_sort(data):

    arr = data.copy()

    comparison = 0
    swap = 0

    n = len(arr)

    for i in range(n):

        for j in range(0, n - i - 1):

            comparison += 1

            if arr[j] > arr[j + 1]:

                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swap += 1

    return arr, comparison, swap


# ==========================================
# Merge Sort
# ==========================================

comparison_merge = 0
swap_merge = 0

def merge_sort(arr):

    global comparison_merge
    global swap_merge

    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2

    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    return merge(left, right)


def merge(left, right):

    global comparison_merge
    global swap_merge

    result = []

    i = 0
    j = 0

    while i < len(left) and j < len(right):

        comparison_merge += 1

        if left[i] < right[j]:

            result.append(left[i])
            i += 1

        else:

            result.append(right[j])
            j += 1

        swap_merge += 1

    while i < len(left):

        result.append(left[i])
        i += 1

    while j < len(right):

        result.append(right[j])
        j += 1

    return result


# ==========================================
# Program Utama
# ==========================================

data = [54, 23, 87, 12, 65, 34, 98, 41, 19, 76]

print("Data Awal")
print(data)

# Bubble Sort

start = time.perf_counter()

bubble_result, bubble_compare, bubble_swap = bubble_sort(data)

bubble_time = (time.perf_counter() - start) * 1000


# Merge Sort

start = time.perf_counter()

merge_result = merge_sort(data.copy())

merge_time = (time.perf_counter() - start) * 1000


# ==========================================
# Output
# ==========================================

print("\n==============================")
print("BUBBLE SORT")
print("==============================")

print("Data :", bubble_result)
print("Comparison :", bubble_compare)
print("Swap :", bubble_swap)
print("Execution Time : %.5f ms" % bubble_time)

print("\n==============================")
print("MERGE SORT")
print("==============================")

print("Data :", merge_result)
print("Comparison :", comparison_merge)
print("Swap :", swap_merge)
print("Execution Time : %.5f ms" % merge_time)
```

Sistem Pencarian Produk Menggunakan Hash Table

```python
# ==========================================
# Class Product
# ==========================================

class Product:

    def __init__(self, code, name):

        self.code = code
        self.name = name

    def __str__(self):

        return f"Kode : {self.code} | Nama : {self.name}"


# ==========================================
# Class HashTable
# ==========================================

class HashTable:

    def __init__(self, size):

        self.size = size

        # setiap index berupa list kosong
        self.table = [[] for _ in range(size)]

    # fungsi hash

    def hash_function(self, key):

        return key % self.size

    # insert

    def insert(self, product):

        index = self.hash_function(product.code)

        self.table[index].append(product)

    # search

    def search(self, code):

        index = self.hash_function(code)

        bucket = self.table[index]

        for product in bucket:

            if product.code == code:

                return product

        return None

    # display

    def display(self):

        print("\nIsi Hash Table")

        print("--------------------------------")

        for i in range(self.size):

            print(f"Index {i} :", end=" ")

            if len(self.table[i]) == 0:

                print("Empty")

            else:

                for product in self.table[i]:

                    print(f"[{product.code}-{product.name}]", end=" ")

                print()


# ==========================================
# Program Utama
# ==========================================

hash_table = HashTable(7)

data_produk = [

    Product(101, "Laptop"),
    Product(102, "Mouse"),
    Product(103, "Keyboard"),
    Product(104, "Monitor"),
    Product(105, "Printer"),
    Product(106, "Speaker"),
    Product(107, "Flashdisk"),
    Product(108, "Webcam")

]

for produk in data_produk:

    hash_table.insert(produk)

hash_table.display()

print()

kode = int(input("Masukkan kode produk : "))

hasil = hash_table.search(kode)

if hasil:

    print("\nProduk ditemukan")

    print("---------------------")

    print("Kode :", hasil.code)

    print("Nama :", hasil.name)

else:

    print("\nProduk tidak ditemukan.")
```

```python
SIZE = 11
hash_table = [None] * SIZE

def hash_function(id_produk):
    return id_produk % SIZE

def tambah_produk():
    id_produk = int(input("ID Produk      : "))
    nama = input("Nama Produk    : ")
    kategori = input("Kategori       : ")
    harga = int(input("Harga          : "))
    terjual = int(input("Jumlah Terjual : "))

    produk = {
        "id": id_produk,
        "nama": nama,
        "kategori": kategori,
        "harga": harga,
        "terjual": terjual
    }

    index = hash_function(id_produk)

    while hash_table[index] is not None:
        index = (index + 1) % SIZE

    hash_table[index] = produk
    print("Produk berhasil ditambahkan!")

def cari_produk():
    id_cari = int(input("Masukkan ID Produk : "))

    index = hash_function(id_cari)
    start = index

    while hash_table[index] is not None:
        if hash_table[index]["id"] == id_cari:
            p = hash_table[index]

            print("\n==========================")
            print("DATA PRODUK")
            print("==========================")
            print("ID          :", p["id"])
            print("Nama        :", p["nama"])
            print("Kategori    :", p["kategori"])
            print("Harga       : Rp{:,.0f}".format(p["harga"]))
            print("Terjual     :", p["terjual"], "unit")
            return

        index = (index + 1) % SIZE

        if index == start:
            break

    print("Produk tidak ditemukan!")

def tampilkan_produk():
    print("\n========================================================")
    print("ID\tNama\t\tKategori\tHarga\t\tTerjual")
    print("========================================================")

    for item in hash_table:
        if item is not None:
            print(
                item["id"],
                "\t",
                item["nama"],
                "\t\t",
                item["kategori"],
                "\t",
                item["harga"],
                "\t",
                item["terjual"]
            )

def quick_sort(data, key, reverse=False):
    if len(data) <= 1:
        return data

    pivot = data[len(data) // 2][key]

    left = []
    middle = []
    right = []

    for item in data:
        if item[key] < pivot:
            left.append(item)
        elif item[key] > pivot:
            right.append(item)
        else:
            middle.append(item)

    if reverse:
        return quick_sort(right, key, True) + middle + quick_sort(left, key, True)

    return quick_sort(left, key) + middle + quick_sort(right, key)

def ranking_penjualan():
    data = [x for x in hash_table if x is not None]

    if not data:
        print("Belum ada data.")
        return

    print("\n1. Terjual (Descending)")
    print("2. Harga (Ascending)")
    print("3. Nama Produk (A-Z)")

    pilih = int(input("Pilih Ranking : "))

    if pilih == 1:
        hasil = quick_sort(data, "terjual", True)
    elif pilih == 2:
        hasil = quick_sort(data, "harga")
    elif pilih == 3:
        hasil = quick_sort(data, "nama")
    else:
        print("Pilihan salah!")
        return

    print("\n========================================================")
    print("ID\tNama\t\tHarga\t\tTerjual")
    print("========================================================")

    for p in hasil:
        print(
            p["id"],
            "\t",
            p["nama"],
            "\t\t",
            p["harga"],
            "\t",
            p["terjual"]
        )

def update_penjualan():
    id_produk = int(input("Masukkan ID Produk : "))

    index = hash_function(id_produk)
    start = index

    while hash_table[index] is not None:
        if hash_table[index]["id"] == id_produk:
            tambah = int(input("Tambah Penjualan : "))
            hash_table[index]["terjual"] += tambah
            print("Data berhasil diperbarui!")
            return

        index = (index + 1) % SIZE

        if index == start:
            break

    print("Produk tidak ditemukan!")

data_awal = [
    {"id": 101, "nama": "Laptop", "kategori": "Elektronik", "harga": 8000000, "terjual": 120},
    {"id": 102, "nama": "Mouse", "kategori": "Aksesoris", "harga": 150000, "terjual": 450},
    {"id": 103, "nama": "Keyboard", "kategori": "Aksesoris", "harga": 300000, "terjual": 350}
]

for p in data_awal:
    idx = hash_function(p["id"])

    while hash_table[idx] is not None:
        idx = (idx + 1) % SIZE

    hash_table[idx] = p

while True:
    print("\n==========================")
    print("ONLINE STORE MANAGEMENT")
    print("==========================")
    print("1. Tambah Produk")
    print("2. Cari Produk")
    print("3. Tampilkan Semua Produk")
    print("4. Ranking Penjualan")
    print("5. Update Penjualan")
    print("6. Keluar")

    pilihan = input("Pilih Menu : ")

    if pilihan == "1":
        tambah_produk()
    elif pilihan == "2":
        cari_produk()
    elif pilihan == "3":
        tampilkan_produk()
    elif pilihan == "4":
        ranking_penjualan()
    elif pilihan == "5":
        update_penjualan()
    elif pilihan == "6":
        print("Program selesai.")
        break
    else:
        print("Pilihan tidak valid!")
```

```python
#   Created by Elshad Karimov
#   Copyright © 2021 AppMillers. All rights reserved.

import math

def bubbleSort(customList):
    for i in range(len(customList)-1):
        for j in range(len(customList)-i-1):
            if customList[j] > customList[j+1]:
                customList[j], customList[j+1] = customList[j+1], customList[j]
    print(customList)


def selectionSort(customList):
    for i in range(len(customList)):
        min_index = i
        for j in range(i+1, len(customList)):
            if customList[min_index] > customList[j]:
                min_index = j
        customList[i], customList[min_index] = customList[min_index], customList[i]
    print(customList)

def insertionSort(customList):
    for i in range(1, len(customList)):
        key = customList[i]
        j = i-1
        while j>=0 and key < customList[j]:
            customList[j+1] = customList[j]
            j -= 1
        customList[j+1] = key
    return customList


def bucketSort(customList):
    numberofBuckets = round(math.sqrt(len(customList)))
    maxValue = max(customList)
    arr = []

    for i in range(numberofBuckets):
        arr.append([])
    for j in customList:
        index_b = math.ceil(j*numberofBuckets/maxValue)
        arr[index_b-1].append(j)

    for i in range(numberofBuckets):
        arr[i] = insertionSort(arr[i])

    k = 0
    for i in range(numberofBuckets):
        for j in range(len(arr[i])):
            customList[k] = arr[i][j]
            k += 1
    return customList

def merge(customList, l, m, r):
    n1 = m - l + 1
    n2 = r - m

    L = [0] * (n1)
    R = [0] * (n2)

    for i in range(0, n1):
        L[i] = customList[l+i]

    for j in range(0, n2):
        R[j] = customList[m+1+j]

    i = 0
    j = 0
    k = l
    while i < n1 and j < n2:
        if L[i] <= R[j]:
            customList[k] = L[i]
            i += 1
        else:
            customList[k] = R[j]
            j += 1
        k += 1
    while i < n1:
        customList[k] = L[i]
        i += 1
        k += 1

    while j < n2:
        customList[k] = R[j]
        j += 1
        k += 1

def mergeSort(customList, l, r):
    if l < r:
        m = (l+(r-1))//2
        mergeSort(customList, l, m)
        mergeSort(customList, m+1, r)
        merge(customList, l, m, r)
    return customList

def partition(customList, low, high):
    i = low - 1
    pivot = customList[high]
    for j in range(low,high):
        if customList[j] <= pivot:
            i += 1
            customList[i], customList[j] = customList[j], customList[i]
    customList[i+1], customList[high] = customList[high], customList[i+1]
    return (i+1)

def quickSort(customList, low, high):
    if low < high:
        pi = partition(customList, low, high)
        quickSort(customList, low, pi-1)
        quickSort(customList, pi+1, high)


def heapify(customList, n, i):
    smallest = i
    l = 2*i + 1
    r = 2*i + 2
    if l < n and customList[l] < customList[smallest]:
        smallest = l

    if r < n and customList[r] < customList[smallest]:
        smallest = r

    if smallest != i:
        customList[i], customList[smallest] = customList[smallest], customList[i]
        heapify(customList, n, smallest)


def heapSort(customList):
    n = len(customList)
    for i in range(int(n/2)-1, -1, -1):
        heapify(customList, n, i)

    for i in range(n-1,0,-1):
        customList[i], customList[0] = customList[0], customList[i]
        heapify(customList, i, 0)
    # customList.reverse()



cList = [2,1,7,6,5,3,4,9,8]
heapSort(cList)
print(cList)
```

```python
def swap(my_list, index1, index2):
    my_list[index1],  my_list[index2] = my_list[index2],  my_list[index1]


def pivot(my_list, pivot_index, end_index):
    swap_index = pivot_index
    for i in range(pivot_index+1, end_index+1):
        if my_list[i] < my_list[pivot_index]:
            swap_index += 1
            swap(my_list, swap_index, i)
    swap(my_list, pivot_index, swap_index)
    return swap_index


def quicksort_helper(my_list, left, right):
    if left < right:
        pivot_index = pivot(my_list, left, right)
        quicksort_helper(my_list, left, pivot_index-1)
        quicksort_helper(my_list, pivot_index+1, right)
    return my_list

def quicksort(my_list):
    return quicksort_helper(my_list, 0, len(my_list)-1)

my_list = [3,5,0,6,2,1,4]
print(quicksort(my_list))
```
