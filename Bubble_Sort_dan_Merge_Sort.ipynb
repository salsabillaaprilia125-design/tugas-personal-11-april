Implementasi Hash Table Sederhana

```python
class HashTable:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size

    # Fungsi Hash
    def hash_function(self, key):
        return key % self.size

    # Insert dengan Linear Probing
    def insert(self, key):
        index = self.hash_function(key)

        while self.table[index] is not None:
            index = (index + 1) % self.size

        self.table[index] = key

    # Menampilkan isi Hash Table
    def display(self):
        print("\nIsi Hash Table")
        print("-" * 30)

        for i in range(self.size):
            if self.table[i] is None:
                print(f"Index {i} : Empty")
            else:
                print(f"Index {i} : {self.table[i]}")


# ===========================
# Program Utama
# ===========================

hash_table = HashTable(10)

data = [25, 36, 47, 58, 69, 80]

for item in data:
    hash_table.insert(item)

hash_table.display()
```

Mengurutkan Data Mahasiswa Menggunakan Quick Sort


```python
# ==================================
# Fungsi Quick Sort
# ==================================

def quick_sort(data):

    if len(data) <= 1:
        return data

    # memilih pivot (elemen tengah)
    pivot = data[len(data) // 2]

    left = []
    middle = []
    right = []

    for item in data:

        if item < pivot:
            left.append(item)

        elif item == pivot:
            middle.append(item)

        else:
            right.append(item)

    return quick_sort(left) + middle + quick_sort(right)


# ==================================
# Program Utama
# ==================================

nilai = [78, 65, 90, 88, 55, 72, 95, 60, 84, 70]

print("Data Awal")
print(nilai)

hasil = quick_sort(nilai)

print("\nData Setelah Quick Sort")
print(hasil)

jumlah_data = len(hasil)
nilai_min = min(hasil)
nilai_max = max(hasil)
rata_rata = sum(hasil) / jumlah_data

print("\nStatistik Data")
print("----------------------")
print("Jumlah Data :", jumlah_data)
print("Nilai Minimum :", nilai_min)
print("Nilai Maksimum :", nilai_max)
print("Nilai Rata-rata :", round(rata_rata, 2))
```

Perbandingan Bubble Sort dan Merge Sort

```python
import time

# ==========================================
# Bubble Sort
# ==========================================

def bubble_sort(data):

    arr = data.copy()

    comparison = 0
    swap = 0

    n = len(arr)

    for i in range(n):

        for j in range(0, n - i - 1):

            comparison += 1

            if arr[j] > arr[j + 1]:

                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swap += 1

    return arr, comparison, swap


# ==========================================
# Merge Sort
# ==========================================

comparison_merge = 0
swap_merge = 0

def merge_sort(arr):

    global comparison_merge
    global swap_merge

    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2

    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    return merge(left, right)


def merge(left, right):

    global comparison_merge
    global swap_merge

    result = []

    i = 0
    j = 0

    while i < len(left) and j < len(right):

        comparison_merge += 1

        if left[i] < right[j]:

            result.append(left[i])
            i += 1

        else:

            result.append(right[j])
            j += 1

        swap_merge += 1

    while i < len(left):

        result.append(left[i])
        i += 1

    while j < len(right):

        result.append(right[j])
        j += 1

    return result


# ==========================================
# Program Utama
# ==========================================

data = [54, 23, 87, 12, 65, 34, 98, 41, 19, 76]

print("Data Awal")
print(data)

# Bubble Sort

start = time.perf_counter()

bubble_result, bubble_compare, bubble_swap = bubble_sort(data)

bubble_time = (time.perf_counter() - start) * 1000


# Merge Sort

start = time.perf_counter()

merge_result = merge_sort(data.copy())

merge_time = (time.perf_counter() - start) * 1000


# ==========================================
# Output
# ==========================================

print("\n==============================")
print("BUBBLE SORT")
print("==============================")

print("Data :", bubble_result)
print("Comparison :", bubble_compare)
print("Swap :", bubble_swap)
print("Execution Time : %.5f ms" % bubble_time)

print("\n==============================")
print("MERGE SORT")
print("==============================")

print("Data :", merge_result)
print("Comparison :", comparison_merge)
print("Swap :", swap_merge)
print("Execution Time : %.5f ms" % merge_time)
```
