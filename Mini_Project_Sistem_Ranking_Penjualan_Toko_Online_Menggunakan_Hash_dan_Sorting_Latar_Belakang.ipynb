Implementasi Hash Table Sederhana

```python
class HashTable:
    def __init__(self, size):
        self.size = size
        self.table = [None] * size

    # Fungsi Hash
    def hash_function(self, key):
        return key % self.size

    # Insert dengan Linear Probing
    def insert(self, key):
        index = self.hash_function(key)

        while self.table[index] is not None:
            index = (index + 1) % self.size

        self.table[index] = key

    # Menampilkan isi Hash Table
    def display(self):
        print("\nIsi Hash Table")
        print("-" * 30)

        for i in range(self.size):
            if self.table[i] is None:
                print(f"Index {i} : Empty")
            else:
                print(f"Index {i} : {self.table[i]}")


# ===========================
# Program Utama
# ===========================

hash_table = HashTable(10)

data = [25, 36, 47, 58, 69, 80]

for item in data:
    hash_table.insert(item)

hash_table.display()
```

Mengurutkan Data Mahasiswa Menggunakan Quick Sort


```python
# ==================================
# Fungsi Quick Sort
# ==================================

def quick_sort(data):

    if len(data) <= 1:
        return data

    # memilih pivot (elemen tengah)
    pivot = data[len(data) // 2]

    left = []
    middle = []
    right = []

    for item in data:

        if item < pivot:
            left.append(item)

        elif item == pivot:
            middle.append(item)

        else:
            right.append(item)

    return quick_sort(left) + middle + quick_sort(right)


# ==================================
# Program Utama
# ==================================

nilai = [78, 65, 90, 88, 55, 72, 95, 60, 84, 70]

print("Data Awal")
print(nilai)

hasil = quick_sort(nilai)

print("\nData Setelah Quick Sort")
print(hasil)

jumlah_data = len(hasil)
nilai_min = min(hasil)
nilai_max = max(hasil)
rata_rata = sum(hasil) / jumlah_data

print("\nStatistik Data")
print("----------------------")
print("Jumlah Data :", jumlah_data)
print("Nilai Minimum :", nilai_min)
print("Nilai Maksimum :", nilai_max)
print("Nilai Rata-rata :", round(rata_rata, 2))
```

Perbandingan Bubble Sort dan Merge Sort

```python
import time

# ==========================================
# Bubble Sort
# ==========================================

def bubble_sort(data):

    arr = data.copy()

    comparison = 0
    swap = 0

    n = len(arr)

    for i in range(n):

        for j in range(0, n - i - 1):

            comparison += 1

            if arr[j] > arr[j + 1]:

                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swap += 1

    return arr, comparison, swap


# ==========================================
# Merge Sort
# ==========================================

comparison_merge = 0
swap_merge = 0

def merge_sort(arr):

    global comparison_merge
    global swap_merge

    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2

    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    return merge(left, right)


def merge(left, right):

    global comparison_merge
    global swap_merge

    result = []

    i = 0
    j = 0

    while i < len(left) and j < len(right):

        comparison_merge += 1

        if left[i] < right[j]:

            result.append(left[i])
            i += 1

        else:

            result.append(right[j])
            j += 1

        swap_merge += 1

    while i < len(left):

        result.append(left[i])
        i += 1

    while j < len(right):

        result.append(right[j])
        j += 1

    return result


# ==========================================
# Program Utama
# ==========================================

data = [54, 23, 87, 12, 65, 34, 98, 41, 19, 76]

print("Data Awal")
print(data)

# Bubble Sort

start = time.perf_counter()

bubble_result, bubble_compare, bubble_swap = bubble_sort(data)

bubble_time = (time.perf_counter() - start) * 1000


# Merge Sort

start = time.perf_counter()

merge_result = merge_sort(data.copy())

merge_time = (time.perf_counter() - start) * 1000


# ==========================================
# Output
# ==========================================

print("\n==============================")
print("BUBBLE SORT")
print("==============================")

print("Data :", bubble_result)
print("Comparison :", bubble_compare)
print("Swap :", bubble_swap)
print("Execution Time : %.5f ms" % bubble_time)

print("\n==============================")
print("MERGE SORT")
print("==============================")

print("Data :", merge_result)
print("Comparison :", comparison_merge)
print("Swap :", swap_merge)
print("Execution Time : %.5f ms" % merge_time)
```

Sistem Pencarian Produk Menggunakan Hash Table

```python
# ==========================================
# Class Product
# ==========================================

class Product:

    def __init__(self, code, name):

        self.code = code
        self.name = name

    def __str__(self):

        return f"Kode : {self.code} | Nama : {self.name}"


# ==========================================
# Class HashTable
# ==========================================

class HashTable:

    def __init__(self, size):

        self.size = size

        # setiap index berupa list kosong
        self.table = [[] for _ in range(size)]

    # fungsi hash

    def hash_function(self, key):

        return key % self.size

    # insert

    def insert(self, product):

        index = self.hash_function(product.code)

        self.table[index].append(product)

    # search

    def search(self, code):

        index = self.hash_function(code)

        bucket = self.table[index]

        for product in bucket:

            if product.code == code:

                return product

        return None

    # display

    def display(self):

        print("\nIsi Hash Table")

        print("--------------------------------")

        for i in range(self.size):

            print(f"Index {i} :", end=" ")

            if len(self.table[i]) == 0:

                print("Empty")

            else:

                for product in self.table[i]:

                    print(f"[{product.code}-{product.name}]", end=" ")

                print()


# ==========================================
# Program Utama
# ==========================================

hash_table = HashTable(7)

data_produk = [

    Product(101, "Laptop"),
    Product(102, "Mouse"),
    Product(103, "Keyboard"),
    Product(104, "Monitor"),
    Product(105, "Printer"),
    Product(106, "Speaker"),
    Product(107, "Flashdisk"),
    Product(108, "Webcam")

]

for produk in data_produk:

    hash_table.insert(produk)

hash_table.display()

print()

kode = int(input("Masukkan kode produk : "))

hasil = hash_table.search(kode)

if hasil:

    print("\nProduk ditemukan")

    print("---------------------")

    print("Kode :", hasil.code)

    print("Nama :", hasil.name)

else:

    print("\nProduk tidak ditemukan.")
```

Mini Project: Sistem Ranking Penjualan Toko Online Menggunakan Hash dan Sorting
Latar Belakang

```python
SIZE = 11
hash_table = [None] * SIZE

def hash_function(id_produk):
    return id_produk % SIZE

def tambah_produk():
    id_produk = int(input("ID Produk      : "))
    nama = input("Nama Produk    : ")
    kategori = input("Kategori       : ")
    harga = int(input("Harga          : "))
    terjual = int(input("Jumlah Terjual : "))

    produk = {
        "id": id_produk,
        "nama": nama,
        "kategori": kategori,
        "harga": harga,
        "terjual": terjual
    }

    index = hash_function(id_produk)

    while hash_table[index] is not None:
        index = (index + 1) % SIZE

    hash_table[index] = produk
    print("Produk berhasil ditambahkan!")

def cari_produk():
    id_cari = int(input("Masukkan ID Produk : "))

    index = hash_function(id_cari)
    start = index

    while hash_table[index] is not None:
        if hash_table[index]["id"] == id_cari:
            p = hash_table[index]

            print("\n==========================")
            print("DATA PRODUK")
            print("==========================")
            print("ID          :", p["id"])
            print("Nama        :", p["nama"])
            print("Kategori    :", p["kategori"])
            print("Harga       : Rp{:,.0f}".format(p["harga"]))
            print("Terjual     :", p["terjual"], "unit")
            return

        index = (index + 1) % SIZE

        if index == start:
            break

    print("Produk tidak ditemukan!")

def tampilkan_produk():
    print("\n========================================================")
    print("ID\tNama\t\tKategori\tHarga\t\tTerjual")
    print("========================================================")

    for item in hash_table:
        if item is not None:
            print(
                item["id"],
                "\t",
                item["nama"],
                "\t\t",
                item["kategori"],
                "\t",
                item["harga"],
                "\t",
                item["terjual"]
            )

def quick_sort(data, key, reverse=False):
    if len(data) <= 1:
        return data

    pivot = data[len(data) // 2][key]

    left = []
    middle = []
    right = []

    for item in data:
        if item[key] < pivot:
            left.append(item)
        elif item[key] > pivot:
            right.append(item)
        else:
            middle.append(item)

    if reverse:
        return quick_sort(right, key, True) + middle + quick_sort(left, key, True)

    return quick_sort(left, key) + middle + quick_sort(right, key)

def ranking_penjualan():
    data = [x for x in hash_table if x is not None]

    if not data:
        print("Belum ada data.")
        return

    print("\n1. Terjual (Descending)")
    print("2. Harga (Ascending)")
    print("3. Nama Produk (A-Z)")

    pilih = int(input("Pilih Ranking : "))

    if pilih == 1:
        hasil = quick_sort(data, "terjual", True)
    elif pilih == 2:
        hasil = quick_sort(data, "harga")
    elif pilih == 3:
        hasil = quick_sort(data, "nama")
    else:
        print("Pilihan salah!")
        return

    print("\n========================================================")
    print("ID\tNama\t\tHarga\t\tTerjual")
    print("========================================================")

    for p in hasil:
        print(
            p["id"],
            "\t",
            p["nama"],
            "\t\t",
            p["harga"],
            "\t",
            p["terjual"]
        )

def update_penjualan():
    id_produk = int(input("Masukkan ID Produk : "))

    index = hash_function(id_produk)
    start = index

    while hash_table[index] is not None:
        if hash_table[index]["id"] == id_produk:
            tambah = int(input("Tambah Penjualan : "))
            hash_table[index]["terjual"] += tambah
            print("Data berhasil diperbarui!")
            return

        index = (index + 1) % SIZE

        if index == start:
            break

    print("Produk tidak ditemukan!")

data_awal = [
    {"id": 101, "nama": "Laptop", "kategori": "Elektronik", "harga": 8000000, "terjual": 120},
    {"id": 102, "nama": "Mouse", "kategori": "Aksesoris", "harga": 150000, "terjual": 450},
    {"id": 103, "nama": "Keyboard", "kategori": "Aksesoris", "harga": 300000, "terjual": 350}
]

for p in data_awal:
    idx = hash_function(p["id"])

    while hash_table[idx] is not None:
        idx = (idx + 1) % SIZE

    hash_table[idx] = p

while True:
    print("\n==========================")
    print("ONLINE STORE MANAGEMENT")
    print("==========================")
    print("1. Tambah Produk")
    print("2. Cari Produk")
    print("3. Tampilkan Semua Produk")
    print("4. Ranking Penjualan")
    print("5. Update Penjualan")
    print("6. Keluar")

    pilihan = input("Pilih Menu : ")

    if pilihan == "1":
        tambah_produk()
    elif pilihan == "2":
        cari_produk()
    elif pilihan == "3":
        tampilkan_produk()
    elif pilihan == "4":
        ranking_penjualan()
    elif pilihan == "5":
        update_penjualan()
    elif pilihan == "6":
        print("Program selesai.")
        break
    else:
        print("Pilihan tidak valid!")
```
